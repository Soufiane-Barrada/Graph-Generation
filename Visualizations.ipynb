# Wave Function Collapse for Graph Generation

## Introduction
In this notebook we aim to generate multiple graphs using the PGG algorithm for different input graphs and of different sizes.

```python
import networkx as nx
import random
from graph_generator import GraphGenerator
from utils import construct_flattened_graph
from stepsVisualization import visualize_all_a_hop_neighborhoods, visualize_wl_labeling, visualize_procedural_generation, visualize_flattened_graph

random.seed(6)
EXTENSION = ".pdf"
```

```python
def generate(graph_generator, size, color_map, colorful=True, save=True, file_name = "foo"):
    # Generate a graph
    generated_tiles= graph_generator.generate_graph(size= size)
    print("Number of nodes:", len(generated_tiles))
    # Flatten the generated graph to a nx.Graph
    flattened_graph= construct_flattened_graph(generated_tiles)
    
    visualize_flattened_graph(flattened_graph, color_map, colorful=colorful, save=save, file_name=file_name)

def img_name(img_num=0, k_hop=0, is_input_graph=False, is_protein=False, size=1):
    file_type = "input" if is_input_graph else "output"
    protein = "_protein" if is_protein else "" 
    ret = "figures/{}_k={}_{}_size={}{}".format(img_num, k_hop, file_type, size, protein) + EXTENSION
    return ret

    
```
